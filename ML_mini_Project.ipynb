```python
column = ['Hobby_top1  ', '  Hobby top2  ', 'Club top1  ', 'Club top 2', '  Music Genre Top 1', '  Game Genre Top1']
df_column = data[column]
display(df_column.head())

for col in df_column.columns:
    print(f"\nUnique values for column: {col}->")
    print(df_column[col].unique())
```

```python
# Calculate frequency of each unique value
value_frequencies = {}
for col in df_column.columns:
    value_counts = df_column[col].value_counts()
    value_frequencies[col] = value_counts

# Analyze top co-occurring pairs
co_occurrence_counts = {}
for index, row in df_column.iterrows():
    items_in_row = []
    for col in df_column.columns:
        item = row[col]
        if pd.notna(item):
            items_in_row.append((col, item))

    for i in range(len(items_in_row)):
        for j in range(i + 1, len(items_in_row)):
            item1_col, item1_val = items_in_row[i]
            item2_col, item2_val = items_in_row[j]

            # Create a unique key for the pair, regardless of order
            pair_key = tuple(sorted(((item1_col, item1_val), (item2_col, item2_val))))

            co_occurrence_counts[pair_key] = co_occurrence_counts.get(pair_key, 0) + 1

# Display most frequent individual items
print("Most Frequent Individual Items:")
for col, frequencies in value_frequencies.items():
    print(f"\nColumn: {col}")
    print(frequencies.head())

# Display most frequent co-occurring pairs
print("\nMost Frequent Co-occurring Pairs:")
sorted_co_occurrences = sorted(co_occurrence_counts.items(), key=lambda item: item[1], reverse=True)
for pair, count in sorted_co_occurrences[:20]: # Display top 20 pairs
    print(f"{pair}: {count}")
```

```python
def recommend_items(student_profile, co_occurrence_counts, value_frequencies):
    """
    Recommends new hobbies and clubs based on a student's profile and co-occurrence counts.
    Args:
        student_profile: A dictionary or Series of the student's existing interests
                         (e.g., {'Hobby_top1  ': 'Cricket', '  Music Genre Top 1': 'Bollywood'}).
        co_occurrence_counts: A dictionary of co-occurrence counts between items.
        value_frequencies: A dictionary of the frequency of each unique value in each column.
    Returns:
        A sorted list of recommended items (hobbies, clubs, music genres, game genres)
        based on their aggregated co-occurrence scores, in descending order.
    """
    recommendation_scores = {}
    student_interests = set()
    for col, item in student_profile.items():
        if pd.notna(item):
            student_interests.add((col, item))

    for (item1_col, item1_val), (item2_col, item2_val) in co_occurrence_counts.keys():
        count = co_occurrence_counts[((item1_col, item1_val), (item2_col, item2_val))]

        if (item1_col, item1_val) in student_interests and (item2_col, item2_val) not in student_interests:
            recommendation_scores[(item2_col, item2_val)] = recommendation_scores.get((item2_col, item2_val), 0) + count
        elif (item2_col, item2_val) in student_interests and (item1_col, item1_val) not in student_interests:
            recommendation_scores[(item1_col, item1_val)] = recommendation_scores.get((item1_col, item1_val), 0) + count

    # Convert recommendations to a list of (item, score) tuples and sort
    sorted_recommendations = sorted(recommendation_scores.items(), key=lambda item: item[1], reverse=True)

    return sorted_recommendations

# Example usage (assuming you have a student_profile, co_occurrence_counts, and value_frequencies)
# Replace with an actual student profile from your data
student_profile_example = df_column.iloc[0].to_dict()
recommendations = recommend_items(student_profile_example, co_occurrence_counts, value_frequencies)
print("For Example.......")
print("\nRecommendations for the first student:")
for rec, score in recommendations:
    print(f"  - {rec[0].strip()} ({rec[1]}): Score {score}")
```

```python
all_recommendations = []

for index, row in df_column.iterrows():
    student_profile = row.to_dict()
    recommendations = recommend_items(student_profile, co_occurrence_counts, value_frequencies)
    # Get the top 10 recommendations
    top_10_recommendations = recommendations[:10]
    all_recommendations.append(top_10_recommendations)

# Create a DataFrame to store the recommendations
recommendations_df = pd.DataFrame({'Recommendations': all_recommendations})

display(recommendations_df.head())
```

```python
import matplotlib.pyplot as plt
from collections import Counter
from mpl_toolkits.mplot3d import Axes3D
import random

# Generate recommendations_df
all_recommendations = []

for index, row in df_column.iterrows():
    student_profile = row.to_dict()
    recommendations = recommend_items(student_profile, co_occurrence_counts, value_frequencies)
    # Get the top 10 recommendations
    top_10_recommendations = recommendations[:10]
    all_recommendations.append(top_10_recommendations)

# Create a DataFrame to store the recommendations
recommendations_df = pd.DataFrame({'Recommendations': all_recommendations})

# Prepare data for 2D visualization (frequency of recommended items)
recommended_items_list = []
for recommendations in recommendations_df['Recommendations']:
    for rec, score in recommendations:
        recommended_items_list.append(f"{rec[0].strip()}: {rec[1]}")

item_counts = Counter(recommended_items_list)
top_items = dict(item_counts.most_common(20)) # Get top 20 most recommended items

plt.figure(figsize=(12, 8))
plt.bar(top_items.keys(), top_items.values())
plt.xticks(rotation=90)
plt.xlabel("Recommended Item")
plt.ylabel("Frequency")
plt.title("Top 20 Most Recommended Hobbies and Clubs")
plt.tight_layout()
plt.show()


# Prepare data for 3D visualization (relationship between top 3 recommendations for each student)
# Extract the top 3 recommendations for each student
top_3_recommendations_per_student = []
for recommendations in recommendations_df['Recommendations']:
    top_3 = [f"{rec[0].strip()}: {rec[1]}" for rec, score in recommendations[:3]]
    # Pad with 'None' if less than 3 recommendations
    while len(top_3) < 3:
        top_3.append('None: None')
    top_3_recommendations_per_student.append(top_3)

# Create a DataFrame for the top 3 recommendations
top_3_df = pd.DataFrame(top_3_recommendations_per_student, columns=['Rec1', 'Rec2', 'Rec3'])

# Assign numerical values to each unique recommendation for plotting
unique_recommendations = sorted(list(set([item for sublist in top_3_recommendations_per_student for item in sublist])))
recommendation_to_int = {item: i for i, item in enumerate(unique_recommendations)}

top_3_df_numeric = top_3_df.replace(recommendation_to_int)

# Create 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

x = top_3_df_numeric['Rec1']
y = top_3_df_numeric['Rec2']
z = top_3_df_numeric['Rec3']

ax.scatter(x, y, z)

ax.set_xlabel('Top 1 Recommendation')
ax.set_ylabel('Top 2 Recommendation')
ax.set_zlabel('Top 3 Recommendation')
ax.set_title('3D Scatter Plot of Top 3 Recommendations per Student')

# Set ticks and labels for better readability
ax.set_xticks(list(recommendation_to_int.values()))
ax.set_xticklabels(list(recommendation_to_int.keys()), rotation=90)
ax.set_yticks(list(recommendation_to_int.values()))
ax.set_yticklabels(list(recommendation_to_int.keys()), rotation=0)
ax.set_zticks(list(recommendation_to_int.values()))
ax.set_zticklabels(list(recommendation_to_int.keys()))

plt.show()
```
